```python
tm=pd.read_csv(r"C:\Users\D\OneDrive\Desktop\abc\titanic\train.csv")
```

```python
# Bar plot for Pclass
plt.figure(figsize=(8,6))
sns.countplot(data=tm, x='Pclass', palette='Reds')
plt.xlabel('Pcalss')
plt.ylabel('Count')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
#Bar plot comparing the number of survivors by Pclass
plt.figure(figsize=(8,6))
sns.countplot(data=tm ,x='Pclass', palette='Greens')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Survivors by Passenger Class')
plt.show()
```

```python
#Correlation heatmap between Age and Fare 
correlation_matrix = tm[['Age' , 'Fare']].corr()
plt.figure(figsize=(9, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues' , fmt='.2f')
plt.title('Correlation Heatmap between Age and Fare')
plt.show()
```
